```python
from PIL import Image
from itertools import product

#conectando ao drive para pegar a imagem
from google.colab import drive
drive.mount('/content/drive')
imagem = Image.open("/content/drive/MyDrive/imgProjetos/harry.jpg")
imagem.show()
```

```python
#Funçaõ para adiconar os tons de cinza
def tons_de_cinza(imagem):
  largura, altura = imagem.size #dimensões da imagem
  img_cinza = Image.new('L', (largura, altura)) #nova imagem criada
  for x, y in product(range(largura), range(altura)):
        r, g, b = imagem.getpixel((x, y))
        # Fórmula de conversão para tons de cinza
        valor_cinza = int(0.299 * r + 0.587 * g + 0.114 * b)
        img_cinza.putpixel((x, y), valor_cinza)
  return img_cinza
```

```python
#Função para fazer a binarização
def binarizacao(img_cinza, limiar = 127): #limiar descreve um nível de exposição
  largura,altura = imagem.size
  img_binaria = Image.new('1', (largura, altura)) #binário
  for x,y in product(range(largura), range(altura)):
      valor_cinza = img_cinza.getpixel((x,y))
      valor_binario = 1 if valor_cinza > limiar else 0
      img_binaria.putpixel((x,y), valor_binario)
  return img_binaria
```

```python
#biblioteca para exibir as imagens diretamente do notebook
from IPython.display import display

display(imagem)

img_cinza = tons_de_cinza(imagem)
display(img_cinza)

img_binaria = binarizacao(img_cinza)
display(img_binaria)

```
